```python
data=df_idf['text'].tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
with open('/content/drive/MyDrive/news.csv','r', encoding='utf-8')as f:
  stopwords=f.readlines()
  stop_set=list(m.strip() for m in stopwords)
  print(stop_set)
```

```python
vectorizer=CountVectorizer(stop_words=stop_set)
word_array=vectorizer.fit_transform(data)
```

```python
print(vectorizer.get_feature_names_out(10))
```

```python
list(vectorizer.get_feature_names_out())[:10]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tf=TfidfTransformer(smooth_idf=True,use_idf=True)
tf.fit(word_array)
```
